```python
plt.scatter(data['Temperature (K)'], data['Absolute magnitude(Mv)'], c=data['Radius(R/Ro)'].astype('category').cat.codes, cmap='rainbow')
plt.style.use('ggplot')
plt.grid()
plt.title("Hertzsprung-Russell Diagram")
plt.xlabel("Absolute Magnitude")
plt.ylabel("Temperature (K)")
```
